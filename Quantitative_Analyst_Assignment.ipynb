## Assignment I

```python
# The task is to get the csv data from the link and concate the data

import pandas as pd
import glob
import os
import zipfile, urllib.request, shutil
stock_name=input('Please enter the stock name (Example, enter:: BTCUSDT ) and then press Enter::')
initial_url = 'https://data.binance.vision/data/spot/daily/trades/BTCUSDT/BTCUSDT-trades-'
initial_url='https://data.binance.vision/data/spot/daily/trades/'+stock_name+'/'+stock_name+'-trades-'

date1= '2022-03-01'
date2= '2022-03-04'
year ='2022'
month = '03'
day1=date1.split('-')[-1]
day2=date2.split('-')[-1]
while(1):
    take_input=input('If you want to enter the dates press 1, else press 0 and click enter (default dates march 1 to march 4)::')
    if take_input=='1':
        year1=input('Enter first day year::')
        month1=input('Enter first day month::')
        day1=input('Enter first day date::')
        year2=input('Enter last day year::')
        month2=input('Enter last day month::')
        day2=input('Enter last day date::')
        date1=str(year1)+'-'+str(month1)+'-'+str(day1)
        date2=str(year2)+'-'+str(month2)+'-'+str(day2)
        break
    elif take_input=='0':
        print('Default start and end dates are::'+date1+' to '+date2)
        break
    else:
        print('Enter either 0 or 1 please::')
        continue

for day in range(int(day1),int(day2)):
    date =year+'-'+month+'-'+'0'+str(day)
    final_url =initial_url+date+'.zip'
    print('Downloading from :: '+ final_url)
   # url = 'http://www....myzipfile.zip'
    file_name = 'myzip.zip'
    url=final_url
    with urllib.request.urlopen(url) as response, open(file_name, 'wb') as out_file:
        shutil.copyfileobj(response, out_file)
        with zipfile.ZipFile(file_name) as zf:
            zf.extractall()

dataframes=[]
for file in glob.glob(os.getcwd()+'/*.csv'):
    print(file)
    df=pd.read_csv(file)
    df.columns=['trade Id' , 'price' ,'qty' ,'quoteQty' ,'time','isBuyerMaker' ,'isBestMatch']# naming the columns
    dataframes.append(df)
    print('Shapes of the files are :: ')
    print(df.shape)
    #myDataFrame = myDataFrame.append(df)
    del df
    os.remove(file)
final_dataframe= pd.concat(dataframes,axis=0) # concatinaing data frame

os.remove(file_name)
final_dataframe.to_csv('Stock_exchange_data_from_'+date1+'_to_'+date2+'_of_'+stock_name+'_stock.csv')
print('All the CSV files of the stock '+stock_name+' for the given dates are combined in the folder where this code runs')
```

#### 1. Bid Size and Offer Size = 1000 USD notional

```python
#Bid Size and Offer Size = 1000 USD notional

# To convert into an equivalent BTC we have devided 1000 USD to the price of BTC in each trade.

final_dataframe['Bid Size'] = 1000/final_dataframe.loc[:,'price']

final_dataframe
```

#### 2. Buy and Sell Margin

```python
# ABSOLUTE VALUE OF BUY AND SELL MARGIN

final_dataframe['Buy and Sell margin'] = abs(final_dataframe.loc[:,'Bid Size'] - final_dataframe.loc[:,'qty'])

final_dataframe
```

#### 3. Maximum position before hedging = 5000 USD notional

```python
final_dataframe['Hedging'] = 5000/final_dataframe.loc[:,'price']

final_dataframe
```

#### Trade match when Possible trading when qty > Bid size

```python
#selecting columns where qty > hedge

dataframe =final_dataframe[final_dataframe.loc[:,'qty'] > final_dataframe.loc[:,'Hedging']]

dataframe
```

```python
#PNLBasis 

dataframe['PNL_Basis'] = (dataframe['Buy and Sell margin']/dataframe['quoteQty'])*10000

#Gross
```

#### Gross PNL in $

```python
dataframe['WIN_PNL'] = dataframe['Buy and Sell margin']*dataframe['price']

dataframe['WIN_PNL'].sum() # Gross PNL in dollars
```

#### Maximum drawdown

```python
# To calculate maximum drawdown we have to calculate first peak value and the lowest vale

# The percentange maximum_drawdown is given as follows:

Maximum_drawdown = ((dataframe['price'].min()-dataframe['price'].max())/dataframe['price'].max())*100

Maximum_drawdown
```
